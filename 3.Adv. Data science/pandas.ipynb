## SERIES

```python
import pandas as pd
```

```python
data = [1,2,3,4,5]
s = pd.Series(data)

print(s)
```

```python
# with dictionary 

dict = {"a":12,"b":13,"c":34}
s = pd.Series(dict)
print(s)
```

```python
#series iwth custom index

col1 =  [1,2,3,4,5]
col2 = ["hihi","huhu","heyhey","haha","htht"]

s = pd.Series(col2,index=col1)
print(s)
```

## DATAFRAME

```python
# 2d means which has rows and columns ... in pandas ,Data frame is a chunk of dataset or can be a entire dataset.
# in pandas , 2d array is called a DATAFRAME
```

```python
a = {
    "Name":["ROhan","Anjali","Jarvis"],
    "class":[11,12,11],
    "age":[22,22,21]
    }
s= pd.DataFrame(a)
print(s)
```

```python
dict =[
    {"name":"Rohan","age":23,"class":"AIDS"},
    {"name":"Robin","age":23,"class":"MBA"},
    {"name":"Angel","age":22,"class":"AIDS"}
]
s = pd.DataFrame(dict)
print(s)
```
