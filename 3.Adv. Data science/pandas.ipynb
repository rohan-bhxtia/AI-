## SERIES

```python
import pandas as pd
```

```python
data = [1,2,3,4,5]
s = pd.Series(data)

print(s)
```

```python
# with dictionary 

dict = {"a":12,"b":13,"c":34}
s = pd.Series(dict)
print(s)
```

```python
#series iwth custom index

col1 =  [1,2,3,4,5]
col2 = ["hihi","huhu","heyhey","haha","htht"]

s = pd.Series(col2,index=col1)
print(s)
```
