## SERIES

```python
import pandas as pd
```

```python
data = [1,2,3,4,5]
s = pd.Series(data)

print(s)
```

```python
# with dictionary 

dict = {"a":12,"b":13,"c":34}
s = pd.Series(dict)
print(s)
```

```python
#series iwth custom index

col1 =  [1,2,3,4,5]
col2 = ["hihi","huhu","heyhey","haha","htht"]

s = pd.Series(col2,index=col1)
print(s)
```

## DATAFRAME

```python
# 2d means which has rows and columns ... in pandas ,Data frame is a chunk of dataset or can be a entire dataset.
# in pandas , 2d array is called a DATAFRAME
```

## .loc ( index based)
### used to locate specific rows or element by index

```python

df.loc[1]
```

```python
df.loc[0:2]  # shows first 3 
```

```python
df.loc[0:2,["Title","IMDB Score"]]   # shows title and imdb score of 0,1,2 indexed rows
```

## iloc ( position based)
##### shows the location regardless of index ... we pass [position], it'll give the position as per rows 

```python
df.iloc[2]     #whatever is present at 3rd row ... bcz we arent working with lables 
```

### helpful when indexes are abc... or something else which will be diffcult for us to call one by one ... then calling positions will give simply rows
